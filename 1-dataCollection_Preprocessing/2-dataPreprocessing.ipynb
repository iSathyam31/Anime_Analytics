### Converting features to JSON format and splitting them into separate columns for better clarity and usability

```python
import json
```

#### For start_season

```python
def split_start_season(data_clean):
    # create NaN columns
    data_clean['start_season_year'] = np.nan
    data_clean['start_season_season'] = np.nan

    for row in range(0, len(data_clean)):
        if pd.isna(data_clean['start_season'][row]):
            continue

        # Ensure the value is a string before trying to load it as JSON
        start_season_value = data_clean['start_season'][row]
        if isinstance(start_season_value, str):
            try:
                start_season = json.loads(start_season_value.replace("'", "\""))
                
                # Check if the loaded value is a dictionary
                if isinstance(start_season, dict):
                    year = start_season.get('year', np.nan)
                    season = start_season.get('season', np.nan)
                else:
                    # Handle case where it's an integer or unexpected type
                    year = np.nan
                    season = np.nan

                # Use .loc[] for assignment to avoid SettingWithCopyWarning
                data_clean.loc[row, 'start_season_year'] = year
                data_clean.loc[row, 'start_season_season'] = season
            except json.JSONDecodeError:
                # Handle the case where the JSON is invalid
                continue
        else:
            # If not a string, treat as invalid or handle differently
            continue
    
    # Drop the original column
    data_clean.drop(columns=['start_season'], inplace=True)

    return data_clean
```

#### For broadcast

```python
def split_broadcast(data_clean):
    # Create new columns with NaN as default
    data_clean['broadcast_day_of_the_week'] = np.nan
    data_clean['broadcast_start_time'] = np.nan

    for index, row in data_clean.iterrows():
        try:
            # Convert the 'broadcast' string to JSON
            broadcast = json.loads(row['broadcast'].replace("'", "\""))
            
            # Assign values to new columns
            data_clean.at[index, 'broadcast_day_of_the_week'] = broadcast.get('day_of_the_week', np.nan)
            data_clean.at[index, 'broadcast_start_time'] = broadcast.get('start_time', 'NIL')
        except (json.JSONDecodeError, KeyError, TypeError):
            # Handle rows with invalid JSON or missing keys
            data_clean.at[index, 'broadcast_day_of_the_week'] = np.nan
            data_clean.at[index, 'broadcast_start_time'] = 'NIL'

    # Drop the original 'broadcast' column
    data_clean.drop(columns=['broadcast'], inplace=True)

    return data_clean
```

#### For Statistics

```python
# Splitting statistics column into watching, completed, on hold, plan to watch and num of user columns

def split_statistics(data_clean):
    # create NaN columns
    data_clean['statistics_watching'] = np.nan
    data_clean['statistics_completed'] = np.nan
    data_clean['statistics_on_hold'] = np.nan
    data_clean['statistics_dropped'] = np.nan
    data_clean['statistics_plan_to_watch'] = np.nan
    data_clean['statistics_num_list_users'] = np.nan

    for row in range(0, len(data_clean)):
        # convert from string to json
        try:
            statistics = json.loads(data_clean['statistics'][row].replace("'", "\""))
            if isinstance(statistics, list):
                statistics = statistics[0]  # Assuming the list contains a dictionary
            # Extract values from the statistics dictionary
            data_clean['statistics_watching'][row] = statistics['status']['watching']
            data_clean['statistics_completed'][row] = statistics['status']['completed']
            data_clean['statistics_on_hold'][row] = statistics['status']['on_hold']
            data_clean['statistics_dropped'][row] = statistics['status']['dropped']
            data_clean['statistics_plan_to_watch'][row] = statistics['status']['plan_to_watch']
            data_clean['statistics_num_list_users'][row] = statistics['num_list_users']
        except Exception as e:
            print(f"Error processing row {row}: {e}")
            continue

    # drop original column
    data_clean.drop(columns=['statistics'], inplace=True)
    
    return data_clean
```

```python
# Create percentage fraction positive/negative viewerships --> Range: [0, 1]
# *function to be called after split_statistics() function*

def get_pos_neg_viewership(anime_index, viewership_types_list, data_clean):
    # Single anime
    total_pos_neg_views = 0
    for viewership_type in viewership_types_list:
        # Ensure the value is numeric (convert to float)
        viewership_value = pd.to_numeric(data_clean.at[anime_index, viewership_type], errors='coerce')
        total_pos_neg_views += viewership_value if pd.notna(viewership_value) else 0
    
    return total_pos_neg_views

def create_viewership_fraction(data_clean):
    # Create NaN columns
    data_clean['positive_viewership_fraction'] = np.nan
    data_clean['negative_viewership_fraction'] = np.nan
    
    positive_viewership = [
        'statistics_watching',
        'statistics_completed',
        'statistics_plan_to_watch'
    ]
    negative_viewership = [
        'statistics_on_hold',
        'statistics_dropped'
    ]
    
    # Loop through rows by index
    for anime_index in range(len(data_clean)):
        total_views = pd.to_numeric(data_clean.at[anime_index, 'statistics_num_list_users'], errors='coerce')
        
        # Check if total views is valid
        if pd.notna(total_views) and total_views > 0:
            # Calculating the total positive and total negative views respectively
            total_pos_views = get_pos_neg_viewership(anime_index, positive_viewership, data_clean)
            total_neg_views = get_pos_neg_viewership(anime_index, negative_viewership, data_clean)
            
            # Calculate percentage fraction & create a new column
            data_clean.at[anime_index, 'positive_viewership_fraction'] = round(total_pos_views / total_views, 4)
            data_clean.at[anime_index, 'negative_viewership_fraction'] = round(total_neg_views / total_views, 4)
    
    return data_clean
```

#### For Studios

```python
def json_studios(data_clean):
    # Convert studios string to json
    for row in range(0, len(data_clean)):
        studio_data = data_clean['studios'][row]
        
        if not studio_data:  # Skip empty or None values
            data_clean['studios'][row] = None
            continue
        
        if isinstance(studio_data, list):  # If it's already a list, no need to parse
            data_clean['studios'][row] = studio_data
            continue
        
        try:
            studios = json.loads(studio_data.replace("'", "\""))
        except json.JSONDecodeError:
            try:
                # Attempt to fix potential common formatting errors
                studios = json.loads(studio_data.replace("'", "\"").replace("\"s", "'s").replace('N\"', "N\'"))
            except json.JSONDecodeError as e:
                print(f"JSONDecodeError at row {row}: {studio_data}")  # Log the error and data for debugging
                data_clean['studios'][row] = None
                continue  # Continue with the next row
        
        data_clean['studios'][row] = studios
    
    return data_clean
```

```python
### FOR THE REVIEWS COLUMN AS THE DATA IS FLOAT
def json_studios_float(data_clean):
    # Convert studios string to json
    for row in range(0, len(data_clean)):
        studio_data = data_clean['studios'][row]
        
        if not studio_data:  # Skip empty or None values
            data_clean['studios'][row] = None
            continue
        
        if isinstance(studio_data, list):  # If it's already a list, no need to parse
            data_clean['studios'][row] = studio_data
            continue
        
        # Ensure studio_data is a string before attempting to replace
        if isinstance(studio_data, str):
            try:
                studios = json.loads(studio_data.replace("'", "\""))
            except json.JSONDecodeError:
                try:
                    # Attempt to fix potential common formatting errors
                    studios = json.loads(studio_data.replace("'", "\"").replace("\"s", "'s").replace('N\"', "N\'"))
                except json.JSONDecodeError as e:
                    print(f"JSONDecodeError at row {row}: {studio_data}")  # Log the error and data for debugging
                    data_clean['studios'][row] = None
                    continue  # Continue with the next row
        else:
            data_clean['studios'][row] = None  # Handle non-string values, like floats, as None
        
        data_clean['studios'][row] = studios
    
    return data_clean
```

#### For genre

```python
def json_genres(data_clean):
    # Convert genres string to json
    for row in range(0, len(data_clean)):
        genre_str = data_clean['genres'][row]
        
        # If it's a list, convert it directly
        if isinstance(genre_str, list):
            data_clean['genres'][row] = genre_str
        elif isinstance(genre_str, str):
            # If it's a string, replace single quotes with double quotes and convert to JSON
            genre_str = genre_str.replace("'", "\"")
            # Check for empty strings or invalid JSON format
            if genre_str.strip():
                try:
                    genres = json.loads(genre_str)
                    data_clean['genres'][row] = genres
                except json.JSONDecodeError:
                    print(f"Invalid JSON format at row {row}: {genre_str}")
            else:
                print(f"Empty string at row {row}")
        else:
            print(f"Unexpected type at row {row}: {type(genre_str)}")
                
    return data_clean
```

#### Cleaning and Preprocess Function

```python
# convert to json and split features
def json_split_features(data_clean):
    # Some datasets do not contain broadcast column
    try:
        data_clean = split_broadcast(data_clean)
    except:
        pass
    
    data_clean = split_start_season(data_clean)
    data_clean = split_statistics(data_clean)
    
    data_clean = create_viewership_fraction(data_clean)
    
    data_clean = json_studios(data_clean)
    data_clean = json_genres(data_clean)
    
    return data_clean
```

```python
# convert to json and split features
def json_split_features_float(data_clean):
    # Some datasets do not contain broadcast column
    try:
        data_clean = split_broadcast(data_clean)
    except:
        pass
    
    data_clean = split_start_season(data_clean)
    data_clean = split_statistics(data_clean)
    
    data_clean = create_viewership_fraction(data_clean)
    
    data_clean = json_studios_float(data_clean)
    data_clean = json_genres(data_clean)
    
    return data_clean
```

## Clean and process rankings csv files

### Function to fill NaN and drop columns

```python
# Filling og NaN value and dropping of unused features
def fill_drop_data(data_clean):
    
    # Drop columns: main_picture, created_at, updated_at, background
    data_clean = data_clean.drop(columns=['main_picture', 'created_at', 'updated_at', 'background'])
    
    # No synopsis information available
    data_clean["synopsis"].fillna(value = "no_synopsis", inplace = True)

    # Anime still airing/ongoing
    data_clean["end_date"].fillna(value = "airing", inplace = True)

    # Anime not broadcasted, replace missing value with the same formet for easy sorting
    # Some dataframes do not contain broadcast column
    try:
        data_clean["broadcast"].fillna(value = "{'day_of_the_week': 'NIL', 'start_time': 'NIL'}", inplace = True)
    except:
        pass
        
    # Source not known
    data_clean["source"].fillna(value = "unknown", inplace = True)

    # Not rated
    data_clean["rating"].fillna(value = "no_rating", inplace = True)

    # No genre information
    data_clean["genres"].fillna(value = "[{'id': -1, 'name': 'no_genre'}]", inplace = True)
    
    # No start season information, so replace year with value -1 and season NIL
    data_clean["start_season"].fillna(value = "{'year': -1, 'season': 'NIL'}", inplace = True)

    # Animes that do not have enough user giving their scorings, so replace null with value -1
    data_clean["mean"].fillna(value = "-1", inplace = True)
    
    # Rank data not available, so replace null with value -1
    data_clean["rank"].fillna(value = "-1", inplace = True)
    
    return data_clean
```

### Iteratively clean and process rankings csv files

```python
for path_num in range(1, len(file_path)):
    # read csv dataset into dataframe
    dataset = pd.read_csv(file_path[path_num])
    
    # cleaning and preprocessing
    data_clean = fill_drop_data(dataset)
    data_clean = json_split_features_float(data_clean)
    
    # create cleaned csv file
    new_file_path = export_df_to_csv(data_clean, file_path[path_num])
    
    print('Done cleaning and preprocessing:', new_file_path)
```

```python
import json  # Needed to parse JSON-like strings if necessary

print('Number of rows before:', len(genres_time_series_df))

# Filter rows where 'genres' has 'no_genre'
valid_rows = []
for row in range(len(genres_time_series_df)):
    try:
        # Parse the genres column if it's a JSON-like string
        genres_data = genres_time_series_df['genres'][row]
        if isinstance(genres_data, str):
            genres_data = json.loads(genres_data)
        
        # Check if the first element has 'name' as 'no_genre'
        if genres_data and genres_data[0]['name'] != 'no_genre':
            valid_rows.append(row)
    except (TypeError, KeyError, IndexError, json.JSONDecodeError) as e:
        # Handle rows that cannot be processed
        pass

# Filter the DataFrame to keep only valid rows
genres_time_series_df = genres_time_series_df.iloc[valid_rows].reset_index(drop=True)

print('Number of rows left:', len(genres_time_series_df))
```

```python
# for each anime, split the genres into individual genre, 
# and add as new row with associated start_season_year, 

# Initialize a list to collect new rows
new_rows = []

for row in range(len(genres_time_series_df)):
    start_season_year = genres_time_series_df.loc[row, 'start_season_year']
    
    # Iterate through each genre in the 'genres' column
    for genre in genres_time_series_df.loc[row, 'genres']:
        # Append the new row to the list
        new_rows.append({"start_season_year": start_season_year, "genres": genre})

# Convert the new rows into a DataFrame
new_rows_df = pd.DataFrame(new_rows)

# Replace the original DataFrame with the new one
genres_time_series_df = new_rows_df.reset_index(drop=True)

# Display the updated DataFrame
genres_time_series_df
```

```python
# rename 'genres' as 'genre'
genres_time_series_df.rename({'genres': 'genre'}, axis="columns", inplace=True)

# drop rows with 'Award Winning' genre since it is not really a genre and it only has 3 rows
print(f"Number of rows before removing 'Award Winning': {len(genres_time_series_df)}")
genres_time_series_df.drop(genres_time_series_df[genres_time_series_df.genre == 'Award Winning'].index, inplace=True)
print(f"Number of rows after removing 'Award Winning': {len(genres_time_series_df)}")

genres_time_series_df
```

```python
# Replace dictionaries with string representations
genres_time_series_df['start_season_year'] = genres_time_series_df['start_season_year'].apply(
    lambda x: str(x) if isinstance(x, dict) else x
)
genres_time_series_df['genre'] = genres_time_series_df['genre'].apply(
    lambda x: str(x) if isinstance(x, dict) else x
)
```

```python
# group according to start_season_year and genres
genres_time_series_df = genres_time_series_df.groupby(['start_season_year', 'genre']).size().reset_index().rename(columns={0: 'count'})
```

```python
genres_time_series_df
```

```python
# create a list for genres
genres_list = [genre for genre in set(genres_time_series_df['genre'])]
print('Number of unique genres:', len(genres_list))


# create a new dataframe for as timeseries
column_names = ['Start Season Year']
column_names.extend([genre for genre in genres_list])

new_genres_time_series_df = pd.DataFrame(columns=column_names)
new_genres_time_series_df.head()
```

```python
# for row in test:
for row in range(len(genres_time_series_df)):
    skip = False
    
    single_year_row = {}

    start_season_year = genres_time_series_df['start_season_year'][row]
    
    # skip years earlier than 1999
    if start_season_year < 1999.0:
        continue
    
    # if start season year already exists in the dataframe, just add 
    for year in new_genres_time_series_df['Start Season Year']:
        if start_season_year == year:
            # add to dataframe
            genre = genres_time_series_df['genre'][row]
            genre_count = genres_time_series_df['count'][row]

            new_genres_time_series_df.loc[new_genres_time_series_df['Start Season Year'] == start_season_year, genre] = genre_count
            
            skip = True
            break
        
    if skip:
        continue
    
    single_year_row['Start Season Year'] = [start_season_year]
    
    for genre in genres_list:
        # add to dictionary the start season year and count
        if genre == genres_time_series_df['genre'][row]:
            single_year_row[genre] = [genres_time_series_df['count'][row]]
        else:
            single_year_row[genre] = [0]
    
    # add to dataframe using pd.concat
    new_row_df = pd.DataFrame(single_year_row)
    new_genres_time_series_df = pd.concat([new_genres_time_series_df, new_row_df], ignore_index=True)
```
